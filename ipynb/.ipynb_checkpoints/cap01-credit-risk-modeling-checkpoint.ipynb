### Outliers Detection

```python
# Percentile based outlier detection
def percentile_based_outlier(data, threshold=95):
    
    #define the percentage difference from the extremes
    diff = (100 - threshold)/2.0
    
    #define the minimum and maximum of this feature
    minimum = np.percentile(data, diff)
    #print (minimum)
    
    maximum = np.percentile(data, (100-diff))
    #print (maximum)
    
    #set True the values between these extremes or False otherwise
    result = ((data < minimum) | (data > maximum))
    
    #number of outlier
    tot_out = 0
    for i in result:
        if i == True: 
            tot_out += 1
    print ("The total number of Percentile Based outlier of '{}' is".format(data.name), tot_out)

    return result
    
    
```

```python
#Median Absolute Deviation (MAD)
#modified Z-score
# link: https://www.ibm.com/support/knowledgecenter/en/SSEP7J_11.1.0/com.ibm.swg.ba.cognos.ug_ca_dshb.doc/modified_z.html
#link: http://colingorrie.github.io/outlier-detection.html
def mad_based_outlier(data, threshold=3.5):
    #find the median
    median = data.median()
    
    #find the absolute difference between points and median
    abs_val = np.abs( [(i - median) for i in data] )
    
    #find mad
    mad = np.median(abs_val)
    
    #find z-score modified
    modified_z_score = [0.6745 * (j-median)/mad for j in data]
    
    result = np.abs(modified_z_score) > threshold
    
    #number of outlier
    tot_out = 0
    for i in result:
        if i == True: 
            tot_out += 1
    print ("The total number of MAD outlier of '{}' is".format(data.name), tot_out)
    
    return result
```

```python
#Standard Deviastion (STD)
def std_div(data, threshold=3.0):
    std = data.std()
    
    result = [(i/std) > threshold for i in data]
    
    #number of outlier
    tot_out = 0
    for i in result:
        if i == True: 
            tot_out += 1
    print ("The total number of STD outlier of '{}' is".format(data.name), tot_out)
    
    return result
    
```

**MAJORITY-VOTE OUTLIER DETECTOR:**

```python
def outlierVote(data):
    x = percentile_based_outlier(data)
    y = mad_based_outlier(data)
    z = std_div(data)
    
    tot=0
    
    temp = zip(data.index, x, y, z)
    final = []
    for i in range(len(temp)):
        if temp[i].count(False) >= 2:
            final.append(False)
        else:
            final.append(True)
            tot +=1
    print ("The total number of Voted outlier of '{}' is".format(data.name), tot)        
    return final
#outlierVote(data=train_data.age)
```

### Visualization of outliers

```python
## This function 'plotOutlier' was copied from https://github.com/jalajthanaki/credit-risk-modelling/blob/master/Credit%20Risk%20Analysis.ipynb


def plotOutlier(x):
    fig, axes = plt.subplots(nrows=4)
    for ax, func in zip(axes, [percentile_based_outlier, mad_based_outlier, std_div, outlierVote]):
        sns.distplot(x, ax=ax, rug=True, hist=False)
        outliers = x[func(x)]
        ax.plot(outliers, np.zeros_like(outliers), 'ro', clip_on=False)

    kwargs = dict(y=0.95, x=0.05, ha='left', va='top', size=20)
    axes[0].set_title('Percentile-based Outliers', **kwargs)
    axes[1].set_title('MAD-based Outliers', **kwargs)
    axes[2].set_title('STD-based Outliers', **kwargs)
    axes[3].set_title('Majority vote based Outliers', **kwargs)
    fig.suptitle('Comparing Outlier Tests with n={}'.format(len(x)), size=20)
    fig = plt.gcf()
    fig.set_size_inches(15,10)
```

## Cross-Validation

Here, we are going to use K-fold Cross Validation to track the overfitting problem and model selection.

```python
from sklearn.model_selection import cross_val_score

def cv_score(func, scr, X_train=X, y_train=y, cv=5, verbose=1):
    cvDict = {}
    for f in func:
        cvScore = cross_val_score(f, X_train, y_train, cv=cv, verbose=verbose, scoring=scr)
        cvDict[str(f).split('(')[0]] = [cvScore.mean(), cvScore.std()]
    return cvDict
```

```python
cross_validation = cv_score(func=[knn, log_reg, adaboost, gradboost, rand_forest], scr='roc_auc')
```

```python
cross_validation
```

As we can see, the two best model still are the same (Adaboost and Gradient Boosting). Apart from that, we can see that the mean and standard deviation scores didn't change too mach and that indicates that our two base models are not suffering of overfittting.
